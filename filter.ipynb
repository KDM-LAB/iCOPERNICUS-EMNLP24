```python
import pandas as pd
import json

# Example DataFrame (Replace this with your actual DataFrame)
# data = {
#     'courseID': ['courseID1', 'courseID1', 'courseID2', 'courseID2'],
#     'courseguidelines': ['guideline1', 'guideline1', 'guideline2', 'guideline2'],
#     'studentID': ['student1', 'student2', 'student3', 'student4'],
#     'review': ['Review1', 'Review3', 'Review5', 'Review7'],
#     'rating': [4, 5, 3, 2]
# }

# df = pd.DataFrame(data)

# Function to convert DataFrame to JSON without outermost key
def dataframe_to_json(df,model):
    # Group by courseID and aggregate the data, storing reviews and ratings for each student
    grouped_df = df.groupby(['posnewID', 'News body','Headline']).apply(
        lambda group: {
            'doc_id': group['posnewID'].iloc[0],
            'doc_text': group['News body'].iloc[0],
            'doc_summ': group['Headline'].iloc[0],
            'm_summary_dict': {
                row['userID']: {
                    'user': row['rewrite_titles'],
                     model: row[model]
                }
                for _, row in group.iterrows()
            }
        }
    ).reset_index(drop=True)

    # Convert the grouped DataFrame to JSON
    json_data = grouped_df.to_json(orient='records', indent=2)

    return json_data

# Example usage
# json_data = dataframe_to_json(tempo)
# data_js = json.loads(json_data)

```

```python
from tqdm.auto import tqdm
import nltk
import re
import traceback
# import typer
import ujson
from tqdm.auto import tqdm
# import utils
# from egises.egises import Summary, Document
import concurrent
import itertools
import os
import typing
from typing import Iterable
import ujson
def _tokenize(text):
    stopwords = set(nltk.corpus.stopwords.words('english'))

    # wordnet lemmatizer
    lemmatizer = nltk.stem.WordNetLemmatizer()

    text = re.sub(r'[^\w\s]', '', text)  # remove punctuation

    text = re.sub(r'[\d+]', '', text.lower())  # remove numerical values and convert to lower case

    tokens = nltk.word_tokenize(text)  # tokenization

    tokens = [token for token in tokens if token not in stopwords]  # removing stopwords

    tokens = [lemmatizer.lemmatize(token) for token in tokens]  # lemmatization

    # my_string= " ".join(tokens)

    return tokens


def _tokenize_text(text):
    return " ".join(_tokenize(text))
def tokenize_experiment_data(input_path: str = f"temp.jsonl",
                             output_filepath: str = "final_Mistralv2_tokenized_consolidated_data.jsonl"):
    """
    input_path: path to consolidated_data.jsonl
    output_filepath: path to finalutils._tokenized_consolidated_data.jsonl
    """

    with open(output_filepath, "w") as fpw:
        with open(input_path, "r") as fp:
            for line in tqdm(fp.readlines()):
                line = line.strip()
                line = ujson.loads(line)
                line["doc_text"] = " ".join(_tokenize(line["doc_text"]))
                line["doc_summ"] = " ".join(_tokenize(line["doc_summ"]))
                new_summ_dict = {}
                for uid, s_dict in line["m_summary_dict"].items():
                    new_summ_dict[uid] = {}
                    for model, summary in s_dict.items():
                        new_summ_dict[uid][model] = " ".join(_tokenize(summary))
                line["m_summary_dict"] = new_summ_dict
                fpw.write(f"{ujson.dumps(line)}\n")
```
